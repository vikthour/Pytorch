# 5 Interesting Pytorch Functions

Pytorch coontains data structures for multi-dimensional tensors and mathematical operations over these are defined.

```python
#import pytorch
import numpy as np
import torch
```

```python
t1 = torch.tensor(4.)
t2 = torch.tensor([1., 2., 4, 6])

t3 = torch.tensor([[1, 4, 5.],
                 [9, 33, 12],
                 [4, 10, 6.]])

t4 = torch.tensor([[[1, 2, 4.],
                    [5, 7, 3]],
                   
                   [[4, 5., 6],
                    [2, 4, 7]]])
```

Function 1

- Torch.diag: If input is a vector (1-D tensor), then returns a 2-D square tensor

```python
'''Example 1 - Generating the diagonal of a 2D square tensor from a vector using the principal diagonal. 
0 is the default positional.'''

x = torch.diag(t2)
x
```

```python
#Example 2 
'''Generating the diagonal of a 2D square tensor from a vector using the the second diagonal. The second diagonal 
is to the right of the normal diagonal'''

y = torch.diag(t2, 1)
y
```

```python
#Example 3
'''Generating the diagonal of a 2D square tensor from a vector from  a matrox using the second diagonal with 
the output set to another tensor.'''

z = torch.diag(t3, 1, out= 3)
z
```

```python
#The output must be set to a tensor not an integer or float otherwise it wouldn't generate an output.
'''It is noteworthy that the output has been streamlined to a vector and the output reflected that the second negative
diagonal was chosen as the output values. The output changes as the diagonal is changed.'''

z = torch.diag(t3, -1, out= (t2))
z
```

Function 2:

- torch.unique: Returns the unique scalar elements of the input tensor.

```python
#Example 1
'''The scalar elements of the input tensor is ouputed based on the uniqueness of the elements'''

x = torch.unique(t2)
x
```

```python
#Example 2
'''The tensor can be sorted in ascending order.'''

y = torch.unique(t3, sorted=True)
y
```

```python
#Example 3
'''The tensor can be sorted and inverted in the output.'''

z = torch.unique(t3/t2, sorted=True, return_inverse=True)
z
```

```python
#Example 3
'''The tensor can be sorted with the output inverted. The previous example has a division between two tensor of different
dimensions hence the error. Meanwhile, if the tensor was divided by a singular tensor, there will be a valid output.
The unique function returns the position of the elements in the original tensor.'''

z = torch.unique(t3/t1, sorted=True, return_inverse=True)
z
```

Function 3

- Torch.round
Returns a new tensor with each of the elements of input rounded to the closest integer.

```python
#Example 1
'''This returns the rounded whole number value of the tensor.'''

x = torch.round(t2 + torch.tensor(0.61))
x
```

```python
#Example 2
'''This returns the rounded value of the matrix tensor.'''

y = torch.round(t3 + torch.tensor(0.61))
y
```

```python
#Example 3
'''The round function does not take an output denoted with an integer.'''

z = torch.round(t4 + torch.tensor(0.61), out=5)
z
```

```python
z = torch.round(t4 + torch.tensor(0.61), out=t1)
z
```

- torch.chunk: Splits a tensor into a specific number of chunks.

```python
#Example 1

x = torch.chunk(t1, t2)
x
```

```python
#Example 2
'''Here the tensor is chunked by the three rows in the matrix.'''

y = torch.chunk(t3, chunks = 3)
y
```

```python
#Example 3
'''The tensor was regarded as a single entity and that's why it can only be chunked into one.'''

z = torch.chunk(t4, chunks = 3)
z
```

```python
class InsuranceModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(input_size, output_size)                  # fill this (hint: use input_size & output_size defined above)
        
    def forward(self, xb):
        out = self.linear(xb)                         # fill this
        return out
    
    def training_step(self, batch):
        inputs, targets = batch 
        # Generate predictions
        out = self(inputs)          
        # Calcuate loss
        loss = F.mse_loss(out, targets)                          # fill this
        return loss
    
    def validation_step(self, batch):
        inputs, targets = batch
        # Generate predictions
        out = self(inputs)
        # Calculate loss
        loss = F.mse_loss(out, targets)                           # fill this    
        return {'val_loss': loss.detach()}
        
    def validation_epoch_end(self, outputs):
        batch_losses = [x['val_loss'] for x in outputs]
        epoch_loss = torch.stack(batch_losses).mean()   # Combine losses
        return {'val_loss': epoch_loss.item()}
    
    def epoch_end(self, epoch, result, num_epochs):
        # Print result every 20th epoch
        if (epoch+1) % 20 == 0 or epoch == num_epochs-1:
            print("Epoch [{}], val_loss: {:.4f}".format(epoch+1, result['val_loss']))
```

```python
def evaluate(model, val_loader):
    outputs = [model.validation_step(batch) for batch in val_loader]
    return model.validation_epoch_end(outputs)

def fit(epochs, lr, model, train_loader, val_loader, opt_func=torch.optim.SGD):
    history = []
    optimizer = opt_func(model.parameters(), lr)
    for epoch in range(epochs):
        # Training Phase 
        for batch in train_loader:
            loss = model.training_step(batch)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
        # Validation phase
        result = evaluate(model, val_loader)
        model.epoch_end(epoch, result, epochs)
        history.append(result)
    return history
```

```python
def accuracy(outputs, labels):
    _, preds = torch.max(outputs, dim=1)
    return torch.tensor(torch.sum(preds == labels).item() / len(preds))
```

```python
class ImageClassificationBase(nn.Module):
    def training_step(self, batch):
        images, labels = batch 
        out = self(images)                  # Generate predictions
        loss = F.cross_entropy(out, labels) # Calculate loss
        return loss
    
    def validation_step(self, batch):
        images, labels = batch 
        out = self(images)                    # Generate predictions
        loss = F.cross_entropy(out, labels)   # Calculate loss
        acc = accuracy(out, labels)           # Calculate accuracy
        return {'val_loss': loss.detach(), 'val_acc': acc}
        
    def validation_epoch_end(self, outputs):
        batch_losses = [x['val_loss'] for x in outputs]
        epoch_loss = torch.stack(batch_losses).mean()   # Combine losses
        batch_accs = [x['val_acc'] for x in outputs]
        epoch_acc = torch.stack(batch_accs).mean()      # Combine accuracies
        return {'val_loss': epoch_loss.item(), 'val_acc': epoch_acc.item()}
    
    def epoch_end(self, epoch, result):
        print("Epoch [{}], val_loss: {:.4f}, val_acc: {:.4f}".format(epoch, result['val_loss'], result['val_acc']))
```

```python
def evaluate(model, val_loader):
    outputs = [model.validation_step(batch) for batch in val_loader]
    return model.validation_epoch_end(outputs)

def fit(epochs, lr, model, train_loader, val_loader, opt_func=torch.optim.SGD):
    history = []
    optimizer = opt_func(model.parameters(), lr)
    for epoch in range(epochs):
        # Training Phase 
        for batch in train_loader:
            loss = model.training_step(batch)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
        # Validation phase
        result = evaluate(model, val_loader)
        model.epoch_end(epoch, result)
        history.append(result)
    return history
```

```python
def get_default_device():
    """Pick GPU if available, else CPU"""
    if torch.cuda.is_available():
        return torch.device('cuda')
    else:
        return torch.device('cpu')
```

```python
device = get_default_device()
device
```

```python
def to_device(data, device):
    """Move tensor(s) to chosen device"""
    if isinstance(data, (list,tuple)):
        return [to_device(x, device) for x in data]
    return data.to(device, non_blocking=True)

class DeviceDataLoader():
    """Wrap a dataloader to move data to a device"""
    def __init__(self, dl, device):
        self.dl = dl
        self.device = device
        
    def __iter__(self):
        """Yield a batch of data after moving it to device"""
        for b in self.dl: 
            yield to_device(b, self.device)

    def __len__(self):
        """Number of batches"""
        return len(self.dl)
```

```python
def plot_losses(history):
    losses = [x['val_loss'] for x in history]
    plt.plot(losses, '-x')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.title('Loss vs. No. of epochs');
```

```python
def plot_accuracies(history):
    accuracies = [x['val_acc'] for x in history]
    plt.plot(accuracies, '-x')
    plt.xlabel('epoch')
    plt.ylabel('accuracy')
    plt.title('Accuracy vs. No. of epochs');
```

```python
input_size = 3*32*32
hidden_size1 = 512
hidden_size2 = 256 
hidden_size3 = 128
hidden_size4 = 64
#hidden_size = 32
output_size = 10
batch_size = 128
```

```python
class CIFAR10Model(ImageClassificationBase):
    def __init__(self, input_size, hidden_size1, hidden_size2, hidden_size3, hidden_size4, output_size):
        super().__init__()
        self.linear1 = nn.Linear(input_size, hidden_size1)
        self.linear2 = nn.Linear(hidden_size1, hidden_size2)
        self.linear3 = nn.Linear(hidden_size2, hidden_size3)
        self.linear4 = nn.Linear(hidden_size3, hidden_size4)
        self.linear5 = nn.Linear(hidden_size4, output_size)
        
    def forward(self, xb):
        # Flatten images into vectors
        out = xb.view(xb.size(0), -1)
        # Apply layers & activation functions
        out = self.linear1(out)
        out = F.relu(out)
        out = self.linear2(out)
        out = F.relu(out)
        out = self.linear3(out)
        out = F.relu(out)
        out = self.linear4(out)
        out = F.relu(out)
        out = self.linear5(out)
        return out
```
